# Sales Data Analysis using Python

```python
import pandas as pd
import matplotlib.pyplot as plt

# read file
data = pd.read_csv("sales.csv")

# show data
print(data.head())

# analysis
sales_by_product = data.groupby("Product")["Sales"].sum()
print(sales_by_product)

# graph
sales_by_product.plot(kind="bar")
plt.title("Sales by Product")
plt.xlabel("Product")
plt.ylabel("Sales")
plt.show()
```

```python
sales_by_city = data.groupby("City")["Sales"].sum()
print(sales_by_city)

sales_by_city.plot(kind="bar")
plt.title("Sales by City")
plt.show()
```

## Insights

- Product B has the highest sales (750)
- Product C has the lowest performance (600)
- Mumbai city contributes the highest sales
- Delhi is second highest in sales
- Sales distribution is not equal across cities and products
